```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import seaborn as sns

%matplotlib inline
```

```python
supp_fig_6bc = pd.read_excel('Suppl_Fig.6bc.xlsx', engine='openpyxl', index_col=0, skiprows=[0])
```

```python
supp_fig_6bc.columns = ['aligner', 'caller', 
                        'noBQSR_total', 'noBQSR_fscore', 'noBQSR_recall', 'noBQSR_precision', 
                        'BQSR_total',   'BQSR_fscore',   'BQSR_recall',   'BQSR_precision']
```

```python
supp_fig_6bc.caller.unique()
```

```python
mutect_noBQSR_recall     = supp_fig_6bc[(supp_fig_6bc.caller=='muTect2')]['noBQSR_recall']
mutect_noBQSR_precision  = supp_fig_6bc[(supp_fig_6bc.caller=='muTect2')]['noBQSR_precision']
strelka_noBQSR_recall    = supp_fig_6bc[(supp_fig_6bc.caller=='strelka')]['noBQSR_recall']
strelka_noBQSR_precision = supp_fig_6bc[(supp_fig_6bc.caller=='strelka')]['noBQSR_precision']
sniper_noBQSR_recall     = supp_fig_6bc[(supp_fig_6bc.caller=='somaticSniper')]['noBQSR_recall']
sniper_noBQSR_precision  = supp_fig_6bc[(supp_fig_6bc.caller=='somaticSniper')]['noBQSR_precision']

mutect_BQSR_recall     = supp_fig_6bc[(supp_fig_6bc.caller=='muTect2')]['BQSR_recall']
mutect_BQSR_precision  = supp_fig_6bc[(supp_fig_6bc.caller=='muTect2')]['BQSR_precision']
strelka_BQSR_recall    = supp_fig_6bc[(supp_fig_6bc.caller=='strelka')]['BQSR_recall']
strelka_BQSR_precision = supp_fig_6bc[(supp_fig_6bc.caller=='strelka')]['BQSR_precision']
sniper_BQSR_recall     = supp_fig_6bc[(supp_fig_6bc.caller=='somaticSniper')]['BQSR_recall']
sniper_BQSR_precision  = supp_fig_6bc[(supp_fig_6bc.caller=='somaticSniper')]['BQSR_precision']
```

```python
fig = plt.figure(figsize=(7, 6))
msize = 60

ax = fig.add_subplot(111)
ax.scatter(mutect_noBQSR_recall,  mutect_noBQSR_precision,  color='tab:red',   s=msize)
ax.scatter(strelka_noBQSR_recall, strelka_noBQSR_precision, color='tab:green', s=msize)
ax.scatter(sniper_noBQSR_recall,  sniper_noBQSR_precision,  color='tab:blue',  s=msize)

ax.set_xlabel('Recall', size=12)
ax.set_ylabel('Precision', size=12)
ax.legend(('MuTect2', 'Strelka2', 'SomaticSniper'), fontsize=12)
_ = plt.yticks( size=12 )
_ = plt.xticks( size=12 )

_ = plt.savefig('../Extended_Data_Figs/ED_Fig_6b.pdf')
```
