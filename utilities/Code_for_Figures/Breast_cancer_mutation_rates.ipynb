```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
n_variants = 41482
```

```python
data = pd.read_csv('Suppl_Data_Files/SNV_indels_in_560_breast_cancers.txt', sep='\t')
```

```python
TNBC = data[~data['TNBC'].isna()]['TNBC'].astype('int')
nonTNBC = data[~data['non-TNBC'].isna()]['non-TNBC']
```

```python
data.head(2)
```

```python
fig = plt.figure(figsize=(9,6))

props = dict(boxes="DarkGreen", whiskers="DarkOrange", medians="DarkBlue", caps="Gray")

colors = ['blue', 'red']

_ = data.boxplot(column=['TNBC', 'non-TNBC',], fontsize=12, patch_artist=True)

_ = plt.scatter(1, n_variants, marker='o', color='red', s=60)
_ = plt.text(0.6, 55000, 'HCC1395', size=12, color='red')
_ = plt.arrow(0.75, 54000, 0.23, n_variants-55000+1500, head_width=0.05, head_length=0.05, color='red')
_ = plt.ylim((0, 60000))
```
